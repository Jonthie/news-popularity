```python
import numpy as np
import csv
import pandas as pd
from sklearn import metrics
from sklearn.preprocessing import StandardScaler,MinMaxScaler, RobustScaler
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.set_option("display.max_columns",100)
import warnings
warnings.simplefilter('ignore')
```

```python
file_path = 'OnlineNewsPopularity.csv'
```

```python
data = pd.read_csv(file_path,delimiter=',')
data.columns = data.columns.str.lower()
#Remove non-predictive
del data['url']
del data[' timedelta']
attributeNames = data.columns[0:-1].values
```

<div class='alert alert-block alert-success' style="font-weight:bolder">

### Filter outliers

</div>

```python
data2 =data
data = data[(np.abs(stats.zscore(data)) < 4).all(axis=1)]
outliers = data2[~data2.apply(tuple,1).isin(data.apply(tuple,1))]
```

<div class='alert alert-block alert-success' style="font-weight:bolder">

### Choose classifier type

</div>

```python
#classify data
def classifyMulti(data):
    classified_data = []
    limits = []
    sampleSize = int(len(data)/5)
    sorted_data = sorted(data)
    for i in range(1,5):
        print("Limit ", i, " is " ,sorted_data[sampleSize*i])
        limits.append(sorted_data[sampleSize*i])
    
    
    for sample in data:
        if sample < limits[0]:
            classified_data.append(0)
        elif sample >= limits[0] and sample < limits[1] :
            classified_data.append(1)
        elif sample >= limits[1] and sample < limits[2]:
            classified_data.append(2)
        elif sample >= limits[2] and sample < limits[3]:
            classified_data.append(3)
        elif sample >= limits[3]:
            classified_data.append(4)
              
    
    return classified_data
```

```python
#classify data
def classifyBin(data):
    classified_data = []
    threshold = data.median()
    print('Median: ', threshold)
    for sample in data:
        if sample < 3395:
            classified_data.append(0)
        else:
            classified_data.append(1)
    
    return classified_data
```

```python
#Extract features and Output
X = data.iloc[:,0:-1]
Y = data.iloc[:, -1]
classified_y = classifyBin(Y) # Choose bin or multi
```

<div class='alert alert-block alert-success' style="font-weight:bolder">

### Feature Selection

### 1 - Mutual Information
    
### 2 - Pearson Correlation

### 3 - RF Feature Importance

</div>

```python
#MUTUAL INFORMATION
from sklearn.feature_selection import mutual_info_classif
mutual_information = mutual_info_classif(X, classified_y, n_neighbors=7)

#Pair feature with score
feature_score = dict(zip(attributeNames,mutual_information))
top_feat = []
for name in attributeNames:
    top_feat.append((name,feature_score[name]))
    
#Extract top 20 best features
data_lesser = pd.DataFrame()
top = 20
cnt=0
for feat in sorted(top_feat, key=lambda tup: tup[1], reverse=True):
    data_lesser[feat[0]] = data[feat[0]]
    cnt +=1
    if cnt == top: #Nr of best Features
        break;
```

```python
feat_import = pd.DataFrame(sorted(top_feat, key=lambda tup: tup[1], reverse=True), columns=['features','score'])
fig, ax = plt.subplots(figsize=(15,10))
plt.xticks(rotation=90)
fig = sns.barplot(x='features',y='score',data=feat_import)
fig.set(xlabel='', ylabel='Feature Importance')
plt.show()
```

```python
#RandomForest FEATURE IMPORTANCE
# Build a forest and compute the feature importances
forest = RandomForestClassifier(n_estimators=300, max_depth=10, criterion='entropy', random_state=0)

forest.fit(X, classified_y)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Extract features
data_lesser = pd.DataFrame()
top = 20
cnt=1
for f in range(X.shape[1]):
    #print("%d. feature %s (%f)" % (f + 1, attributeNames[indices[f]], importances[indices[f]]))
    data_lesser[attributeNames[indices[f]]] = data[attributeNames[indices[f]]]
    if cnt == top: #Nr of best Features
        break;
    cnt+=1
```

```python
#Robust Scaler
scaler = RobustScaler()
scaler.fit(data_lesser)
data_transformed = scaler.transform(data_lesser)
```

```python
#Normalize data
scaler = MinMaxScaler()
scaler.fit(data_lesser)
data_transformed = scaler.transform(data_lesser)
```

```python
#Standardize data
scaler = StandardScaler()
scaler.fit(data_lesser)
data_transformed = scaler.transform(data_lesser)
```

<div class='alert alert-block alert-success' style="font-weight:bolder">

### Models

### 1 - RandomForest
    
### 2 - K Nearest Neighbor

### 3 - Support Vector Machine

</div>

```python
X_train, X_test, Y_train, Y_test = train_test_split(data_transformed, classified_y, test_size = 0.2, random_state = 0)
```

```python
data_lesser.head()
```

```python
#Cross validation 10-fold
rndForest = RandomForestClassifier(n_estimators=100, max_depth=10, criterion='gini', random_state=0)
score = (cross_val_score(rndForest, data_transformed, classified_y, cv=5))
print("Min score: ",min(score))
print("Max Score",max(score))
print("Mean Score",score.mean())
print("scores: ",score)
```

```python
#Confusion matrix
rndForest.fit(X_train, Y_train)
rFtest = rndForest.predict(X_test)
res = confusion_matrix(Y_test, rFtest)
print(res)
print((res[0][0] + res[1][1])/(sum(res[0])+sum(res[1])))
```

```python
#Find K
accuracy_knn = []
for n in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=n)
    score = (cross_val_score(knn, data_transformed, classified_y, cv=5))
    accuracy_knn.append(score.mean())
    
fig, ax = plt.subplots(figsize=(15,10))
plt.xlabel('Number of Neightbors')
plt.ylabel('Accuracy')
plt.plot(range(1,30), accuracy_knn)
```

```python
knn = KNeighborsClassifier(n_neighbors=15)
score = (cross_val_score(knn, data_transformed, classified_y, cv=5))
print("Min score: ",min(score))
print("Max Score",max(score))
print("Mean Score",score.mean())
print("scores",score)
```

```python
#Confusion matrix
knn.fit(X_train, Y_train)
knn_test = knn.predict(X_test)
res = confusion_matrix(Y_test, knn_test)
print(res)
```

```python
lsvc = LinearSVC(C=0.01, penalty="l2", dual=False, multi_class = 'ovr')
score = (cross_val_score(lsvc, data_transformed, classified_y, cv=5))
print("Min score: ",min(score))
print("Max Score",max(score))
print("Mean Score",score.mean())
print("scores",score)
```

```python
#create images and shares graph
temp_data = data
temp_data['shares'] = Y
max_img = int(max(data[' num_imgs']))
totShares = [0]*(max_img+1)

temp_data = temp_data[[' num_imgs','shares']]
for index, row in temp_data.iterrows():
    totShares[int(row[' num_imgs'])] += row['shares']
    
img_shares = pd.DataFrame()
img_shares['numimg'] = range(max_img+1)
img_shares['shares'] = totShares

fig, ax = plt.subplots(figsize=(15,10))
fig = sns.barplot(x='numimg',y='shares', data=img_shares)
fig.set(xlabel='Numer of Images', ylabel='Total Number of Shares')
plt.show()
```

```python
#create Data channel and shares graph
temp_data = data
temp_data['shares'] = Y
channels = [' data_channel_is_tech',' data_channel_is_socmed',' data_channel_is_bus'
            ,' data_channel_is_world',' data_channel_is_entertainment', ' data_channel_is_lifestyle', 'shares']

temp_data = temp_data[channels]
result_channel = []

for d in channels[0:-1]:
    result_channel.append(sum(temp_data[temp_data[d] == 1]['shares']))
    
chan_share = pd.DataFrame()
chan_share['channel'] = channels[0:-1]
chan_share['shares'] = result_channel

fig, ax = plt.subplots(figsize=(15,10))
fig = sns.barplot(x='channel',y='shares', data=chan_share)
fig.set(xlabel='', ylabel='Number of Shares')
plt.show()
```
